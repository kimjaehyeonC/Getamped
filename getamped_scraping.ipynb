```python
import requests
import time
import re
from bs4 import BeautifulSoup

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt

import warnings
warnings.filterwarnings(action='ignore')

%matplotlib inline
plt.style.use('seaborn')
plt.rc('font', family = 'Malgun Gothic')
```

```python
df = pd.DataFrame(columns = ['액세서리', '외형', '성능', '재미'])
df
```

```python
# 상세 url 페이지에서 세부 평점 데이터 수집
def star_rating(url):
    res2 = requests.get(url)
    bs2 = BeautifulSoup(res2.text, 'html.parser')
    
    tmp = []
    for i in range(3):
        tmp.append(float(bs2.select('div#contents tr li p.star')[i].text))
    
    return tmp
```

```python
res = requests.get('https://getamped.juneinter.com/sub_main/menu/item/view/item_accessory.ws?page=35&srch_type=item_name&srch=&order_type=regdate')
bs = BeautifulSoup(res.text, 'html.parser') # 3 5 10 12 17 19
```

```python
# 페이지 돌면서 수집
j = 1
while True:
    
    res = requests.get(f'https://getamped.juneinter.com/sub_main/menu/item/view/item_accessory.ws?page={j}&srch_type=item_name&srch=&order_type=regdate')
    bs = BeautifulSoup(res.text, 'html.parser') # 3 5 10 12 17 19
    
    # 끝 페이지에 도달하면 종료
    if (j%10 == 1) & (bs.select('div#paging') == []):
        break
    
    #
    n = len(bs.select('table#t_chlist > tr > td ')) # 항목 개수
    t = bs.select('table#t_chlist > tr > td ') # 항목의 내용 불러오기
    for i in range(3, n+1, 7): # td의 개수 7개를 건너 뛰고 다음 항목으로 넘어가기 위한 for loop
        if float(t[i+2].text.strip()) > 0: # 평점 항목에 값이 있다면 크롤링 진행
            if list(df.index):
                df.loc[df.index[-1]+1] = [t[i].text, 0, 0 ,0]
            else:
                df.loc[0] = [t[i].text, 0, 0 ,0]
                
            # 세부 평점 크롤링을 위해 상세페이지로 진입 및 데이터 수집 
            url = 'https://getamped.juneinter.com/sub_main/menu/item/view/' + t[i].a['href']

            df.loc[df.index[-1], ['외형', '성능', '재미']] = star_rating(url)
            
    print(f'--{j}page 완료--')    
    j += 1
    time.sleep(1)
    
```

```python
df.to_csv('C:/datafile/겟앰상점악평점.csv', encoding = 'cp949')
```
